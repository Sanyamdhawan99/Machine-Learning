```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
```

```python
train_data = np.genfromtxt("GDpr1_train.csv", delimiter = ",")
X = train_data[:, 0:13]
Y = train_data[:, 13]
print(X)
print(Y)
```

```python
alg1 = LinearRegression()
alg1.fit(X, Y)
print(alg1.coef_)
print(alg1.intercept_)
```

```python
test_data = np.genfromtxt("GDpr1_test.csv", delimiter = ",")
X_test = test_data[:, 0:13]
print(X_test)
```

```python
Y_test = alg1.predict(X_test)
print(Y_test)
```

```python
np.savetxt("Y_GDProject1_test_predicted.csv", Y_test, fmt = "%0.5f", delimiter=",")
```
