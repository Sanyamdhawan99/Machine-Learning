```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = [1, 2, 3]
y = [4, 5, 6]
plt.plot(x, y, "b-.")
plt.show()
```

```python
# graph for x^4
#x = np.array([1,2,3,4])
x = np.arange(0, 5, 0.1)
y = x ** 4
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graphs")
plt.grid()
plt.axis([0, 10, 0, 1000]) # argument is a list
plt.text(4, 200, "sample text", fontsize = 12)
plt.plot(x, y, color="red", linewidth = 3, label = "x^4")
plt.legend() # compulsory to show the legend
plt.show()
```

```python
# pie graphs
sizes = [10, 20, 30, 40]
colors = ["red", "blue", "green", "grey"]
labels = ["Red", "Blue", "Green", "Grey"]
explode = [0.1, 0, 0, 0]
plt.pie(sizes, colors = colors, labels = labels, autopct = "%0.1f%%", explode = explode, counterclock=False, startangle = 120)
plt.title("Colors")
plt.axis("equal") # makes it a circle
plt.show()
```
