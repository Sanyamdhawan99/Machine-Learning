```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("twitter_train.csv")
test_data = pd.read_csv('twitter_test.csv') 
```

```python
df = data.copy()
df_test = test_data.copy()
```

```python
x_train = df['text']
y_train = df['airline_sentiment']
x_test = df_test['text']
```

```python
# importing regular expression
import re
x_train = [re.sub('@[^\s]+', '', x) for x in x_train]
x_train
```

```python
x_test = [re.sub('@[^\s]+', '', x) for x in x_test]
x_test
```

```python
from nltk.tokenize import word_tokenize
```

```python
x_train = [word_tokenize(x) for x in x_train]
x_train
```

```python
x_test = [word_tokenize(x) for x in x_test]
x_test
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords
stops = stopwords.words('english')
import string
punctuations = list(string.punctuation)
stops += punctuations
stops
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk import pos_tag
def clean_tweet(words):
    cleaned_output = []
    for w in words:
        words_check.append(w)
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))
            cleaned_output.append(clean_word.lower())
    return cleaned_output 
```

```python
x_train = [clean_tweet(i) for i in x_train]
x_train
```

```python
x_test = [clean_tweet(i) for i in x_test]
x_test
```

```python
training_data = [" ".join(x) for x in x_train]
training_data
```

```python
testing_data = [" ".join(x) for x in x_test]
testing_data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
count_vec = CountVectorizer(max_features=3000, ngram_range=(1,2))
```

```python
x_train_features = count_vec.fit_transform(training_data)
x_train_features
```

```python
x_test_features = count_vec.transform(testing_data)
x_test_features
```
