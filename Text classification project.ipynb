```python
# include punctuations in stop words
import string
punctuations = list(string.punctuation)
stop_words += punctuations
stop_words
```

```python
#fetch data from the files 
path = "E:/Machine Learning/20_newsgroups/"
category = glob.glob(path + '*')
print(category)
for i in range(len(category)):
    category[i] = category[i].replace(path, '')
all_documents = []
all_documents_category = []
for i in range(len(category)):
    path_to_folder = category[i] + "/"
    text_files = glob.glob(path_to_folder + '*')
    print(path_to_folder)
    for document in text_files:
        f = open(document, "rb")
        all_documents.append(f.read().decode(errors='replace'))
        all_documents_category.append(category[i])
```

```python
df = pd.DataFrame()
df["text_data"] = all_documents
df["target"] = all_documents_category
# shuffle DataFrame
df = df.reindex(np.random.permutation(df.index))
df.reset_index(drop = True)
```

```python
#test train split
X = df.text_data
Y = df.target
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, random_state = 1) 
```

```python
# features selection using training data
dictionary = {} 

for doc in X_train:
    words_in_doc = doc.split()
    for word in words_in_doc:
        if(word in stop_words):
            continue
        elif(word in dictionary.keys()): 
            dictionary[word] = dictionary[word] + 1
        else: 
            dictionary[word] = 1
        
        
vocab = dictionary        
vocab = sorted(vocab.items(), key=lambda x: x[1], reverse=True)
feature_set = vocab[0:2000]
feature_set
```

```python
# voc = final feature set 
voc = []
for i in range(len(feature_set)):
    voc.append(feature_set[i][0])
voc
```

# Implementing multinomial naive bayes

```python
def fit(x_train_, y_train_):
    result = {}
    all_classes = set(y_train_)
    num_features = x_train_.shape[1] 
    result["total_documents"] = len(y_train_)
    for current_class in all_classes :
        result[current_class] = {}
        valid_rows = (y_train_ == current_class)
        x_current_rows = x_train_[valid_rows]
        y_current_rows = y_train_[valid_rows]
        result[current_class]["total_doc_with_current_class"] = len(y_current_rows) 
        count_all_words = 0
        for j in range(1, num_features + 1):
            result[current_class][j] = x_current_rows[:, j - 1].sum()
            count_all_words += result[current_class][j]
        result[current_class]["total_words"] = count_all_words    
        
    return result
```

```python
def predict(dictionary, x_test_):
    y_pred = []
    for x in x_test_:
        class_pred = predictSinglePoint(dictionary, x) 
        y_pred.append(class_pred)
    return y_pred    
```

```python
def predictSinglePoint(dictionary, x):
    all_classes = dictionary.keys()
    best_p = -1
    best_class = -1 
    first_run = True
    for current_class in all_classes :
        if current_class == "total_documents":
            continue
        p_current_class = probability(dictionary, x, current_class)
        if(first_run or p_current_class > best_p):
            best_class = current_class
            best_p = p_current_class
        first_run = False   
    return best_class    
```

```python
def probability(dictionary, x, current_class):
    output = np.log(dictionary[current_class]["total_doc_with_current_class"]) - np.log(dictionary["total_documents"])
    num_features = len(dictionary[current_class].keys()) - 2
    for j in range(1, num_features + 1):
        if x[j - 1] != 0:
            feature_count = dictionary[current_class][j] + 1
            all_words_count = dictionary[current_class]["total_words"] + num_features 
            res = np.log(feature_count) - np.log(all_words_count)
            output = output + res
    return output
```

```python
# implementation of classifier 
dict = fit(x_train_, y_train_)
y_pred_2 = predict(dict, x_test_)
print(classification_report(y_pred_2, y_test_))
print(confusion_matrix(y_pred_2, y_test_))
```
